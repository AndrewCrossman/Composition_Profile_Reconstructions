## Resequence File Names with timestamp

```python
directory_path = '/content/drive/Shareddrives/BNL_June_2024/DataFiles/Cu1Ni2_200C/integration'
tth_files = glob.glob(directory_path + '/*.xy')

date_format = '%Y%m%d%H%M%S'
time_sorter = pd.DataFrame(columns=['datetime','seq_num'])

for file in tth_files:
  file_number = re.findall(r'\d+',file)[7:]
  date_obj = datetime.datetime.strptime(file_number[0]+file_number[1], date_format)
  time_sorter.loc[len(time_sorter.index)] = [date_obj,float(file_number[-1])]

time_sorter.sort_values(['datetime','seq_num']).reset_index( drop=True)
```
