```python
import csv
import numpy as np
import pandas as pd
import scipy as sp
from scipy import optimize
from google.colab import files
import matplotlib.pylab as plt
%pylab inline
```

```python
def gauss(x,mu,sigma):
  return ( 1/(np.sqrt(2*np.pi*sigma*sigma)) * np.exp(-((x-mu)*(x-mu))/(2*sigma*sigma)) )
```

```python
def half_gauss(x,mu,sigma):
  return ( 1/(np.sqrt(2*np.pi*sigma*sigma)) * np.exp(-((x-mu)*(x-mu))/(2*sigma*sigma)) )
```

```python
mu = 0
sigma = .5
bin_size = .02
atoms = 1000 # per bin
#x = np.arange(0,2+bin_size,bin_size)
x = np.arange(-2-bin_size,2+bin_size,bin_size)
y = [int(np.random.binomial(atoms, half_gauss(i,mu,sigma), 1))/atoms for i in x]

fig1, ax1 = plt.subplots(figsize=(6,4,),dpi=200)
ax1.bar(x,y,width=bin_size, label="Measured")
ax1.plot(x, [half_gauss(i,mu,sigma) for i in x], color='red', label="Expected" )
ax1.legend(fontsize=16, bbox_to_anchor=[1,.75], loc='center right')
ax1.set_title("1000 Atoms Per Layer", fontsize=18, y=.9)
ax1.set_ylabel("Concentration %A", fontsize=18)
ax1.set_xlabel("Distance", fontsize=18)
#ax1.set_xlim(0,2)
ax1.set_xlim(0-2,2)
ax1.set_ylim(0,1.2)
ax1.set_xticks(ticks=[-2,-1,0,1,2],labels=[-1,-.5,0,.5,1],fontsize=14)
ax1.set_yticks(ticks=[0,.2,.4,.6,.8,1],labels=[0,.2,.4,.6,.8,1],fontsize=14)
plt.tight_layout()
```
