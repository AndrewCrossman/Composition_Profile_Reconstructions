```python
import numpy as np
import pandas as pd
import scipy as sp
from sklearn.neighbors import KDTree
import matplotlib.pylab as plt
%pylab inline
```

```python
def LJ(E0,R0):
  LJ_Table = {}
  R03 = R0*R0*R0
  R06 = R03*R03
  R012 = R06*R06
  A = E0*R012
  B = 2*E0*R06
  for r in np.linspace(2,10,100):
    r3 = r*r*r
    r6 = r3*r3
    r12 = r6*r6
    LJ_Table[r] = A/r12 -B/r6
  return LJ_Table
```
